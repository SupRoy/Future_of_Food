```python
# Reading the dataset of Meat Consumption into a dataframe

Meat = pd.read_csv("Meat Production and Consumption/OECD_Outlook_2021_Meat_Europe.csv")
Meat.head()
```

```python
# Dropping unnecessary columns

Meat.drop(['DATAFLOW', 'CONVENTION', 'FREQ'], axis= 1, inplace= True)
Meat.head()
```

```python
Meat.drop(Meat[Meat['REF_AREA']!='E'].index, inplace= True) # Dropping the countries to only keep the data for Europe
Meat.drop(Meat[Meat['MEASURE'] != 'QC'].index, inplace= True) # Keeeping only data for consumption
Meat.head()
```

```python
## Extracting Pork from the Meat Consumption Data

Pork = copy.deepcopy(Meat)
Pork.drop(Pork[Pork['COMMODITY'] != 'CPC_EX_PK'].index, inplace= True) # Keeping only data for pigmeat or pork
Pork.rename(columns={"OBS_VALUE":"Pork"},inplace= True)
Pork.drop(['REF_AREA', 'COMMODITY', 'MEASURE', 'UNIT_MEASURE', 'VERSION_ID', 'OBS_STATUS', 'UNIT_MULT', 'DECIMALS'], axis= 1, inplace= True)
Pork.head()
```

```python
## Extracting Beef and Veal from the Meat Consumption Data

Beef = copy.deepcopy(Meat)
Beef.drop(Beef[Beef['COMMODITY'] != 'CPC_EX_BV'].index, inplace= True) # Keeping only data for beef and veal
Beef.rename(columns={"OBS_VALUE":"Beef_and_veal"},inplace= True)
Beef.drop(['REF_AREA', 'COMMODITY', 'MEASURE', 'UNIT_MEASURE', 'VERSION_ID', 'OBS_STATUS', 'UNIT_MULT', 'DECIMALS'], axis= 1, inplace= True)
Beef.head()
```

```python
## Extracting Poultry from the Meat Consumption Data

Poultry = copy.deepcopy(Meat)
Poultry.drop(Poultry[Poultry['COMMODITY'] != 'CPC_EX_PT'].index, inplace= True) # Keeping only data for poultry
Poultry.rename(columns={"OBS_VALUE":"Poultry"},inplace= True)
Poultry.drop(['REF_AREA', 'COMMODITY', 'MEASURE', 'UNIT_MEASURE', 'VERSION_ID', 'OBS_STATUS', 'UNIT_MULT', 'DECIMALS'], axis= 1, inplace= True)
Poultry.head()
```

```python
## Extracting Sheep meat from the Meat Consumption Data

Sheep = copy.deepcopy(Meat)
Sheep.drop(Sheep[Sheep['COMMODITY'] != 'CPC_EX_SH'].index, inplace= True) # Keeping only data for sheep meat
Sheep.rename(columns={"OBS_VALUE":"Sheep"},inplace= True)
Sheep.drop(['REF_AREA', 'COMMODITY', 'MEASURE', 'UNIT_MEASURE', 'VERSION_ID', 'OBS_STATUS', 'UNIT_MULT', 'DECIMALS'], axis= 1, inplace= True)
Sheep.head()
```

```python
## Merging all meat consumptions into a dataframe (Unit: Thousand Tonnes, Region: Europe)

df = pd.merge(pd.merge(pd.merge(Pork, Beef, on= 'TIME_PERIOD'), Poultry, on='TIME_PERIOD'), Sheep, on='TIME_PERIOD')
df.rename(columns={"TIME_PERIOD":"Year"}, inplace= True)
df.head()
```

```python
## Importing Population dataset

Population = pd.read_csv("Meat Production and Consumption/FAOSTAT_European_Population_1990_2021.csv")
Population.drop(Population[Population['Element Code'] != 511].index, inplace = True) #Keeping only the data for Total Population
Population.drop(Population[Population['Area'] == 'Holy See'].index, inplace = True)
Population.drop(Population[Population['Area'] == 'Gibraltar'].index, inplace = True)
Population.drop(Population[Population['Area'] == 'Isle of Man'].index, inplace = True)
Population.drop(Population[Population['Area'] == 'Liechtenstein'].index, inplace = True)
Population.head()
```

```python
# Checking the Countries present in the Population Dataset

Population['Area'].unique()
```

```python
Pop = Population.groupby(['Year']).Value.sum().reset_index() #Extracting Population for each year from 1990 to 2021
Pop
```

```python
# Merging Population (in thousands) with meat consumption in Europe

df = pd.merge(df, Pop, on= "Year")
df.rename(columns={"Value":"Population"}, inplace= True)
df.head()
```
